```python
from keras.models import Sequential
from keras.layers import Dense,Activation
import numpy
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from keras.optimizers import SGD

X, y = make_moons(500, noise=0.20)
plt.scatter(X[:,0], X[:,1], s=40, c=y, cmap=plt.cm.Spectral)

# create model
model = Sequential()
model.add(Dense(90, input_dim=2, kernel_initializer='normal', activation='relu'))
model.add(Dense(40, activation='relu', kernel_initializer='normal'))
model.add(Dense(20, activation='relu', kernel_initializer='normal'))
model.add(Dense(1, activation='sigmoid', kernel_initializer='normal'))
# Compile model
model.compile(loss='mean_squared_error', optimizer=SGD(lr=0.5), metrics=['accuracy'])
# Fit the model
model.fit(X, y, epochs=100)
# Scores
scores = model.evaluate(X, y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# calculate predictions
predictions = model.predict(X)


plt.show()
```

Se observa que la precision en el caso de solo una neurona es de 87%, lo que es demasiado alto para un problema tan simple, por lo tanto se necesitan mas capas y neuronas para poder lograr una separacion del dataset. 

Por otro lado, para el caso multicapa se obtiene una precision del 97% la que podria ser mejor utilizando otra arquitectura, sin embargo, es una mejora en considerable en comparacion del caso de una neurona.
